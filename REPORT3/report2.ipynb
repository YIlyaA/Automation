## KONFIGURACJA

```python
import os
import main
from matplotlib import pyplot as plt
import numpy as np

TEST_SIZES = [100, 5000, 10000, 15000, 20000, 25000, 30000, 35000, 40000, 45000, 50000, 55000, 60000, 65000, 70000]
MAX_NUM = 1000000
MIN_NUM = 10                            
SOURCES = 'sources'
BINS = 'bins'
TESTS = 'tests'
RESULTS = 'results'
VERBOSE = False
```

## TWORZENIE TESTÓW

```python
main.create_tests(TESTS, TEST_SIZES, MAX_NUM, MIN_NUM)
```

## KOMPILACJA

```python
main.compile_sources(SOURCES, BINS, v=True)
```

## URUCHOMIENIE

```python
for algo in os.listdir(BINS):
    for ts in TEST_SIZES:
        main.run_algo(BINS, TESTS, RESULTS, ts, algo, v=VERBOSE)
```

```python
create, minim, balance = main.read_results(RESULTS)

plt.title("TWORZENIE STRUKTURY", fontsize=16, color='green')
main.plot_graf(create, marker='.', linestyle='-', undtitle=False)

plt.title("WYSZUKANIE MINIMUM", fontsize=16, color='green')
main.plot_graf(minim, marker='.', linestyle='-', undtitle=True)

plt.title("RÓWNOWAŻENIE", fontsize=16, color='green')
main.plot_graf(balance, marker='.', linestyle='-', undtitle=False)
```
