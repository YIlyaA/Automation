## KONFIGURACJA

```python
import os
import main
from matplotlib import pyplot as plt
import numpy as np

TEST_SIZES = [10, 15, 20, 25, 20, 25, 30, 35, 40]
MAX_NUM = 100
MIN_NUM = 10                            
SOURCES = 'sources'
BINS = 'bins'
TESTS = 'tests'
RESULTS = 'results'
VERBOSE = False
```

## TWORZENIE TESTÓW

```python
main.create_tests(TESTS, TEST_SIZES, MAX_NUM, MIN_NUM)
```

## KOMPILACJA

```python
main.compile_sources(SOURCES, BINS, v=True)
```

## URUCHOMIENIE

```python
for algo in os.listdir(BINS):
    for ts in TEST_SIZES:
        main.run_algo(BINS, TESTS, RESULTS, ts, algo, v=VERBOSE)
```

```python
create, minim, balance = main.read_results(RESULTS)

plt.title("TWORZENIE STRUKTURY", fontsize=16, color='green')
main.plot_graf(create, marker='.', linestyle='-')
plt.title("WYSZUKANIE MINIMUM", fontsize=16, color='green')
main.plot_graf(minim, marker='.', linestyle='--')
plt.title("RÓWNOWAŻENIE", fontsize=16, color='green')
main.plot_graf(balance, marker='.', linestyle='-')
```
